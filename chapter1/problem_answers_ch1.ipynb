```python
# 1.1

# The ciphertext below was encrypted using a substitution cipher.
# Decrypt the ciphertext without knowledge of the key.


cipher_text = """lrvmnir bpr sumvbwvr jx bpr lmiwv yjeryrkbi jx qmbm wi
bpr xjvni mkd ymibrut jx irhx wi bpr riirkvr jx
ymbinlmtmipw utn qmumbr dj w ipmhh but bj rhnvwdmbr bpr
yjeryrkbi jx bpr qmbm mvvjudwko bj yt wkbrusurbmbwjk
lmird jk xjubt trmui jx ibndt
wb wi kjb mk rmit bmiq bj rashmwk rmvp yjeryrkb mkd wbi
iwokwxwvmkvr mkd ijyr ynib urymwk nkrashmwkrd bj ower m
vjyshrbr rashmkmbwjk jkr cjnhd pmer bj lr fnmhwxwrd mkd
wkiswurd bj invp mk rabrkb bpmb pr vjnhd urmvp bpr ibmbr
jx rkhwopbrkrd ywkd vmsmlhr jx urvjokwgwko ijnkdhrii
ijnkd mkd ipmsrhrii ipmsr w dj kjb drry ytirhx bpr xwkmh
mnbpjuwbt lnb yt rasruwrkvr cwbp qmbm pmi hrxb kj djnlb
bpmb bpr xjhhjcwko wi bpr sujsru msshwvmbwjk mkd
wkbrusurbmbwjk w jxxru yt bprjuwri wk bpr pjsr bpmb bpr
riirkvr jx jqwkmcmk qmumbr cwhh urymwk wkbmvb"""

# Key space = 26! â‰ˆ 2^88
# bruteforcing will take too long
# so lets do letter frequency analysis

# 1) computer relative frequency of all letters in the cipher text

letter_counts = {}

# remove the whitespace and \n newline
removed_newline_cipher_text = cipher_text.replace("\n", "")
removed_space_cipher_text = removed_newline_cipher_text.replace(" ", "")


# counting the letters
for letter in removed_space_cipher_text:
    if letter in letter_counts:
        letter_counts[letter] += 1
    else:
        letter_counts[letter] = 1

# calculatng relative frequencies, and rounding to 4 decimal places
cipher_relative_frequencies = {letter: round(count / len(removed_space_cipher_text), 4) for letter,
                        count in letter_counts.items()}

# sorting relative frequencies by highest to lowest frequency
sorted_cipher_relative_frequencies = dict(sorted(cipher_relative_frequencies.items(),
                                          key=lambda item: item[1],
                                          reverse=True))

print("Cipher text sorted by decreasing relative frequencies:")
print(sorted_cipher_relative_frequencies)
print("")

# 2) decrypt cipher text with help of english letter frequency table

# English letter frequency table

english_relative_frequencies = {
    'a': 0.0817,
    'b': 0.0150,
    'c': 0.0278,
    'd': 0.0425,
    'e': 0.1270,
    'f': 0.0223,
    'g': 0.0202,
    'h': 0.0609,
    'i': 0.0697,
    'j': 0.0015,
    'k': 0.0077,
    'l': 0.0403,
    'm': 0.0241,
    'n': 0.0675,
    'o': 0.0751,
    'p': 0.0193,
    'q': 0.0010,
    'r': 0.0599,
    's': 0.0633,
    't': 0.0906,
    'u': 0.0276,
    'v': 0.0098,
    'w': 0.0236,
    'x': 0.0015,
    'y': 0.0197,
    'z': 0.0007
}

# sort from high to low
sorted_english_relative_frequencies = dict(sorted(english_relative_frequencies.items(),
                                                 key=lambda item: item[1],
                                                 reverse=True))

print("English language sorted by decreasing relative frequency:")
print(sorted_english_relative_frequencies)
print("")

# lets now subsitute our sorted letter frequencies to our cipher text

decrypted_cipher = ""
for letter in cipher_text:
    if letter.isalpha(): # if its an alphanumerical char, so we dont replace " "
        # find the index of the letter in the cipher frequency table
        index = list(sorted_cipher_relative_frequencies.keys()).index(letter)
        # find the corresponding letter in the english frequency table
        decrypted_letter = list(sorted_english_relative_frequencies.keys())[index]
        decrypted_cipher += decrypted_letter
    else: # if its a space we just add the space letter
        decrypted_cipher += letter

print("First pass decrypted cipher:")
print("")
print(decrypted_cipher)
print("")

# we see that its not perfect, so lets manually subsitute some letters

decrypted_cipher = decrypted_cipher.replace("y", "b")
decrypted_cipher = decrypted_cipher.replace("x", "q")
decrypted_cipher = decrypted_cipher.replace("k", "x")
decrypted_cipher = decrypted_cipher.replace("p", "k")
decrypted_cipher = decrypted_cipher.replace("f", "p")
decrypted_cipher = decrypted_cipher.replace("u", "f")
decrypted_cipher = decrypted_cipher.replace("w", "u")
decrypted_cipher = decrypted_cipher.replace("j", "w")
decrypted_cipher = decrypted_cipher.replace("o", "n")
decrypted_cipher = decrypted_cipher.replace("i", "o")
decrypted_cipher = decrypted_cipher.replace("g", "y")

print("Manually cleaned decrypted cipher:")
print("")
print(decrypted_cipher)
print("")

# some subsitutions cause loops, notable, replacing n with i

# 3) Who wrote the text?
# A quick google search shows the text was written by
print(" - Shoshin Nagamine, the founder of Matsubayashi-ryu Karate.")
```

```python
# 1.2

# We received the following ciphertext which was encoded with a shift cipher:

cipher_text = "xultpaajcxitltlxaarpjhtiwtgxktghidhipxciwtvgtpilpitghlxiwiwtxgqadds"

# 1) Perform an attack based on letter frequency count
# I first had problems understanding the question
# but we want to use the frequency of the letters, to find the shift value

letter_counts = {}

# counting the letters
for letter in cipher_text:
    if letter in letter_counts:
        letter_counts[letter] += 1
    else:
        letter_counts[letter] = 1

# calculatng relative frequencies, and rounding to 4 decimal places
cipher_relative_frequencies = {letter: round(count / len(cipher_text), 4) for letter,
                        count in letter_counts.items()}

# sorting relative frequencies by highest to lowest frequency
sorted_cipher_relative_frequencies = dict(sorted(cipher_relative_frequencies.items(),
                                          key=lambda item: item[1],
                                          reverse=True))

print(sorted_cipher_relative_frequencies)
print("")

# we see that the most common letter is t
# so our shift value, from t -> e, is 11

# so we only have to recover one letter, to crack the shift cipher

# at first i thought 14, from e->t
# but we want to overlap
# so we go from t->z->e, which is 11

print("Encrypted shift cipher:")
print(cipher_text)
print("")

# also will be trying to use functions from now on
# instead of one-shotting each task

# to solve, i will create a function which takes an input string, and shift value
# then making a 'ring' set, which contains the alphabet a-z, and the order values
# i then will iter over the input string
# turning the string into an array of values
# by comparing the array to the ring
# then, add the shift value to each char in the input string array
# and if the value overflows 26, loop the value back to 1
# and finally map the ring to the value array
# and turn it back into a string

def shiftDecrypter(input_string, shiftValue):
    
    # for future ref
    # modulo is much easier with 0 based indexing
    possible_chars = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4,
    'e': 5,
    'f': 6,
    'g': 7,
    'h': 8,
    'i': 9,
    'j': 10,
    'k': 11,
    'l': 12,
    'm': 13,
    'n': 14,
    'o': 15,
    'p': 16,
    'q': 17,
    'r': 18,
    's': 19,
    't': 20,
    'u': 21,
    'v': 22,
    'w': 23,
    'x': 24,
    'y': 25,
    'z': 26
}
    input_string_value_array = []
    
    for letter in input_string:
        input_string_value_array.append(possible_chars[letter])
        
    for index, value in enumerate(input_string_value_array):
        
        #cuz 1 based indexing we can only do modulo when value > 26
        if (value + shiftValue) > 26:
            shifted_value = (value + shiftValue) % 26
        else:
            shifted_value = value + shiftValue
            
        input_string_value_array[index] = shifted_value
        
    decrypted_text = ""
    
    for value in input_string_value_array:
        for key, dict_value in possible_chars.items():
            if value == dict_value:
                decrypted_text += key
        
    return decrypted_text

print("Decrypted shift cipher:")
print(shiftDecrypter(cipher_text, 11))
print(" - Tecumseh")
```
